# Read in creditcardfraud_clean.csv from previous file location

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

#Set up directories first 
#CreditCardFraud
#CreditCardFraud/features0
#CreditCardFraud/labels0
#CreditCardFraud/feature1
#CreditCardFrayd/labels1

data_path = "SET\\YOUR\\PATH\\CreditCardFraud\\"

#header None to remove the first row of heade
data = pd.read_csv(data_path + "creditcardfraud_clean.csv",header = None)

data.head(n = 10)
```

# Scale Column 29 into range of [-1, 1]

```python
data[28] = StandardScaler().fit_transform(data[28].values.reshape(-1, 1))

data.shape
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt

data_subset = data.iloc[0:284807][29]

labels_dist = pd.value_counts(data_subset.iloc[:], sort = True)

labels_dist.plot(kind = 'bar', rot = 0)
plt.title('Transaction Class Distribution')
plt.xticks(range(2), ["Normal", "Fraud"])
plt.xlabel('Class')
plt.ylabel('Frequency')


labels_dist

```

# Save data point into separate csv files(feature & label).  
# Location of the files depending on the label value 
### There are 4 directories: features0, labels0, features1, labels1 

```python

normal = 0 # [0 - 284314] 284315 items 
fraud = 284315 #[284315 - 284806] 492 items 

for i in range(0, data.shape[0]):
    
    current_row_data = data.iloc[[i]]
    current_row_feature = current_row_data[current_row_data.columns[0:29]]
    current_row_label = current_row_data[29]
    
    if current_row_label[i].astype('bool') == False:
        current_row_feature.to_csv(data_path + "features0\\" + str(normal) + ".csv", header = None, index = False)
        current_row_label.to_csv(data_path + "labels0\\" + str(normal) + ".csv", header = None, index = False)
        
        normal = normal + 1
        
    else: #current_row_label == 1
        current_row_feature.to_csv(data_path + "features1\\" + str(fraud) + ".csv", header = None, index = False)
        current_row_label.to_csv(data_path + "labels1\\" + str(fraud) + ".csv", header = None, index = False)
        
        fraud = fraud + 1
    
```
